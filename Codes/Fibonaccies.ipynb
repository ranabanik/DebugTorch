Fibonacci series is a basic problem in computer science. The series goes as 1, 1, 2, 3, 5, 8, 13, ... How many ways to find the value of n<sup>th</sup> member in the series. <br>
Along with coding shrewdness, factors like time and memory efficiency also come into the play. 

```python
import numpy as np
import time
```

My implementation

```python
def fibonacci(n): #-> int
    if type(n) != int:
        raise TypeError("n must be a positive int")
    elif n<1:
        print("Number can't less or equal to 0")
#     elif type(n) == float:
#         print("Must provide integer")
    else:
        L = np.zeros(n+1)
        L[0] = 1
        L[1] = 1
        for i in range(2, n+1):
            L[i] = L[i-1]+L[i-2]
        return int(L[n-1])
```

Time taken: 0.39467334747314453

```python
fibonacci(2.4)
```

```python
n=1000
start = time.time()
for i in range(1, n+1):
    print("{} : {}".format(i, fibonacci(i)))
end = time.time()
print("Time taken: ", end - start)
```

**Memoization**: To store recent values so that the function doesn't have to work.

```python
fibonacci_cache = {}
def fibonacci3(n):
    """
    check to see if the value exists in fibonacci_cache. If exists simply return it.
    other wise compute the value, cache/store it and return.  
    """
    if n in fibonacci_cache:
        return fibonacci_cache[n]
    # compute the nth term
    if n==1:
        value = 1
    elif n==2:
        value = 1
    elif n > 2:
        value = fibonacci3(n-1) + fibonacci3(n-2)
        
    # cache the value and return it
    fibonacci_cache[n]=value
    return value
```

```python
fibonacci3(100)
```
